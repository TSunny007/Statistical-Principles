# The aim of the Birthday Paradox is to measure the expected collision of random events.

First, let us consider the famous situation of birthdays. Lets make formal the setting. Consider a room of k people, chosen at random from the population, and assume each person is equally likely to have any birthday (excluding February 29th), so there are n = 365 possible birthdays.

The probability that any two (i.e. k = 2) people (ALICE and BOB) have the same birthday is 1/n = 1/365 ≈ 0.003. The birthday of ALICE could be anything, but once it is known by ALICE, then BOB has probability 1/365 of matching it.
To measure that at least one pair of people have the same birthday, it is easier to measure the probability that no pair is the same. For k = 2 the answer is 1 − 1/n and for n = 365 that is about 0.997.
For a general number k (say k = 23) there are  ($^k_2$)  = k · (k − 1)/2 (read as k choose 2) pairs. For k = 23,
then  23  = 253. Note that  ($^k_2$)  = Θ($^k_2$). 

We need for each of these events that the birthdays do not match. Assuming independence we have
(1 − 1/n)$^{(^k_2)}$ or 0.997253 = 0.467.
And the probability there is a match is thus 1 minus this number
1 − (1 − 1/n)$^{(^k_2)}$ or 1 − 0.997253 = 0.532,
just over 50%.

#### Really, the probability should be

$ 1 - (\frac{n−1}{n}) * \frac{n−2}{n} * \frac{n−3}{n} * ...=1− \prod_{i=1}^{k-1}(\frac{n-i}{n})$

```python
def birthday(possibilities):
    visited = set()
    trials = 0
    while True:
        nextRan = random.randint(1,possibilities)
        if nextRan not in visited:
            visited.add(nextRan)
            continue
        else: # This is when the collision actually occurs
            break
    return len(visited)+1 # We add one because the collision doesn't reflect on the size of the set
```

```python
n = 365
print('%d trials to get a repeat'%birthday(n))
### expected
print('when the expected amount of trials to get a repeat is %d'%math.sqrt(2*n))
```

Let's perform 300 random trials and plot out the rate of success with varying values of K

```python
y = []
k = []
trials = 300
totalFrac = 1/trials
for q in range(0, trials):
    nextRand = birthday(n)
    k.append(nextRand)
k = sorted(k)
for z in range(len(k)):
    y.append(k.index(k[z])*totalFrac)
```

```python
sns.set_style("darkgrid")
mpl.rcParams['figure.figsize'] = (6,4)
mpl.rcParams['figure.dpi'] = 200
plt.title("collision given k random draws with n = %d possibilities" %n)
plt.xlabel("k")
plt.ylabel("fraction of success")
plt.xlim([0,70])
plt.ylim([-0.1,1.1])
plt.plot(np.array(k), np.array(y))
plt.show()
```
